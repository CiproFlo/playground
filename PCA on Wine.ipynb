```python
#!/usr/bin/env python
%matplotlib inline
#%config InlineBackend.figure_format = 'svg'

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Set random state for debugging only
RANDOM_STATE = 42

# Load data set from sklearn
X, y = datasets.load_wine().data, datasets.load_wine().target

# Scale data to mean = zero and SD = unit variance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=13)
X_reduced = pca.fit_transform(X_scaled)

#print(pca.explained_variance_ratio_)

# Plot the proportion of variance explained (PVE)
plt.figure(figsize=[5, 3], dpi=100)
X_labels = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7',
            'PC8', 'PC9', 'PC10', 'PC11', 'PC12', 'PC13',]
plt.rcParams['axes.facecolor'] = '#fafafa'
plt.title("Proportion of variance explained")
plt.bar(range(pca.explained_variance_ratio_.shape[0]), 
        pca.explained_variance_ratio_, color="#20936d",
        align="center")
plt.xticks(range(pca.explained_variance_ratio_.shape[0]), 
           X_labels, rotation=45, ha='center')
#plt.xlim([-1, X.shape[1]])
plt.show();
```

```python
# Plot the cummulative PVE
plt.figure(figsize=[5, 3], dpi=100)
X_labels = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7',
            'PC8', 'PC9', 'PC10', 'PC11', 'PC12', 'PC13',]
plt.rcParams['axes.facecolor'] = '#fafafa'
plt.title("Proportion of variance explained")
plt.plot(range(pca.explained_variance_ratio_.shape[0]), 
        np.cumsum(pca.explained_variance_ratio_), color="#20936d",
        )
plt.xticks(range(pca.explained_variance_ratio_.shape[0]), 
           X_labels, rotation=45, ha='center')
#plt.xlim([-1, X.shape[1]])
plt.ylim(-0.1,1.1)
plt.show();
```
