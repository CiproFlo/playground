```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
from sklearn import datasets

df = pd.read_csv('_DATA/iris.csv')

plt.figure();
andrews_curves(df, 'Name');
```

```python
from sklearn.decomposition import PCA

# Split df into data to fit and transform and target vector
data = df.drop('Name', axis=1)
species = df['Name']

# PCA maintaining the same shape as input.
estimator = PCA(n_components=len(data.columns))
data_transformed = estimator.fit_transform(data)
data_transformed = pd.DataFrame(data=data_transformed,    
                                index=range(data.shape[0]),    
                                columns=['PCA'+str(i) for i in range(1, len(data.columns)+1)])
data_transformed = pd.concat([data_transformed, df[['Name']]], axis=1)

andrews_curves(data_transformed, 'Name');
```
